```python
import pandas as pd
```

```python
csv_file = 'Resources/disney_plus_titles.csv'
disney = pd.read_csv(csv_file, index_col = False)
disney.head()
```

```python
disney_movies = disney.loc[disney['type']=='Movie']
disney_movies.head()
```

```python
disney_movies = disney_movies[['title','director','cast','country','release_year']]
disney_movies.head()
```

```python
disney_movies.to_csv('Resources/Disney_Movies.csv')
```
