```python
#drop values with invalid rating data
final_df = final_df.drop(final_df[final_df['rating'] == "74 min"].index)
final_df = final_df.drop(final_df[final_df['rating'] == "84 min"].index)
final_df = final_df.drop(final_df[final_df['rating'] == "66 min"].index)
```

```python
# rename the repetitive/similar ratings
final_df['rating'] = final_df['rating'].replace(to_replace = 'TV-G', value = 'G')
final_df['rating'] = final_df['rating'].replace(to_replace = 'TV-Y', value = 'G')
final_df['rating'] = final_df['rating'].replace(to_replace = 'TV-PG', value = 'PG')
final_df['rating'] = final_df['rating'].replace(to_replace = 'TV-7', value = 'PG')
final_df['rating'] = final_df['rating'].replace(to_replace = 'TV-Y7', value = 'PG')
final_df['rating'] = final_df['rating'].replace(to_replace = 'TV-Y7-FV', value = 'PG')
final_df['rating'] = final_df['rating'].replace(to_replace = 'TV-Y7-FX', value = 'PG')
final_df['rating'] = final_df['rating'].replace(to_replace = 'TV-14', value = 'PG-13')
final_df['rating'] = final_df['rating'].replace(to_replace = 'TV-MA', value = 'R')
final_df['rating'] = final_df['rating'].replace(to_replace = 'NC-17', value = 'R')
final_df['rating'] = final_df['rating'].replace(to_replace = 'UR', value = 'NR')
```

```python
# number of movies by rating per platform
print("Count of Ratings on Each Streaming Platform:")
rating_by_platform = final_df.groupby(['rating', 'streaming']).size().reset_index(name="Count").sort_index(ascending=True)
print(rating_by_platform)
```
