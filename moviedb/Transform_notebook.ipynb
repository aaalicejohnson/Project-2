```python
# import libraries
import pandas as pd
```

```python
# read csv files
disney_df = pd.read_csv('disney_api.csv',index_col=0)
netflix_df = pd.read_csv('netflix_API_list.csv',index_col=0)
```

```python
disney_df
```

```python
netflix_df
```

```python
# concatenate the dataframes
final_df = pd.concat([disney_df,netflix_df],axis=0)
final_df.head()
```

```python
#drop values with invalid rating data
final_df = final_df.drop(final_df[final_df['rating'] == "74 min"].index)
final_df = final_df.drop(final_df[final_df['rating'] == "84 min"].index)
final_df = final_df.drop(final_df[final_df['rating'] == "66 min"].index)
```

```python
# rename the repetitive/similar ratings
final_df['rating'] = final_df['rating'].replace(to_replace = 'TV-G', value = 'G')
final_df['rating'] = final_df['rating'].replace(to_replace = 'TV-Y', value = 'G')
final_df['rating'] = final_df['rating'].replace(to_replace = 'TV-PG', value = 'PG')
final_df['rating'] = final_df['rating'].replace(to_replace = 'TV-7', value = 'PG')
final_df['rating'] = final_df['rating'].replace(to_replace = 'TV-Y7', value = 'PG')
final_df['rating'] = final_df['rating'].replace(to_replace = 'TV-Y7-FV', value = 'PG')
final_df['rating'] = final_df['rating'].replace(to_replace = 'TV-Y7-FX', value = 'PG')
final_df['rating'] = final_df['rating'].replace(to_replace = 'TV-14', value = 'PG-13')
final_df['rating'] = final_df['rating'].replace(to_replace = 'TV-MA', value = 'R')
final_df['rating'] = final_df['rating'].replace(to_replace = 'NC-17', value = 'R')
final_df['rating'] = final_df['rating'].replace(to_replace = 'UR', value = 'NR')
print(final_df['rating'].value_counts())
```

```python
final_df
```

```python
# write to csv
final_df.to_csv('all_movies.csv')
```

```python
# number of movies by rating per platform
print("Count of Ratings on Each Streaming Platform:")
rating_by_platform = final_df.groupby(['rating', 'streaming']).size().reset_index(name="Count").sort_index(ascending=True)
print(rating_by_platform)
```

```python
# number of movies by rating per platform
disney_df = final_df[final_df['streaming']=='Disney+']
netlfix_df = final_df[final_df['streaming']=='Netflix']

disney_ratings = [len(disney_df[disney_df['rating']=='G']),
                  len(disney_df[disney_df['rating']=='PG']),
                  len(disney_df[disney_df['rating']=='PG-13']),
                  len(disney_df[disney_df['rating']=='R'])]
netflix_ratings = [len(netflix_df[netflix_df['rating']=='G']),
                   len(netflix_df[netflix_df['rating']=='PG']),
                   len(netflix_df[netflix_df['rating']=='PG-13']),
                   len(netflix_df[netflix_df['rating']=='R'])]
movie_ratings_df = pd.DataFrame(data={'Disney+': disney_ratings, 'Netflix': netflix_ratings},
                                index=['G', 'PG', 'PG-13', 'R'])
movie_ratings_df.plot.bar(title= 'MPAA Ratings on Disney+ & Netflix',
                          color= ['#153866', '#E50914'],
                          width= .8)
```

```python
# number of movies per streaming pltform
movies_by_platform = final_df.groupby('streaming').count()['title']
movies_by_platform
```

```python
# number of movies per streaming platform by year
print("Count of Movies on Each Streaming Platform by Year:")
yearly_count_by_platform = final_df.groupby(['release_year', 'streaming']).size().reset_index(name="Count").sort_index(ascending=True)

print(yearly_count_by_platform)
```

```python
# movies by country
print("Count of Movies by Country:")
movies_by_country = final_df.groupby('country').count()['title']
print(movies_by_country)
```

```python
# Horror Movies
genre_df = final_df.dropna().reset_index(drop=True)
horror_movies=genre_df[genre_df['genre'].str.contains('Horror', case=False)].reset_index(drop=True)
horror_movies.head()
```

```python
# Animation Movies
animation_movies = genre_df[genre_df['genre'].str.contains('Animation', case=False)].reset_index(drop=True)
animation_movies.head()
```

```python
# Comedy Movies
comedy_movies = genre_df[genre_df['genre'].str.contains('Comedy', case=False)].reset_index(drop=True)
comedy_movies.head()
```

```python
horror_movies.to_csv('horror_movies.csv')
animation_movies.to_csv('animation_movies.csv')
comedy_movies.to_csv('comedy_movies.csv')
```
