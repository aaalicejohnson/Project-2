```python
#Dependencies

import pandas as pd
import requests
import json
# import config
from json.decoder import JSONDecodeError
from config import api_key
from pprint import pprint
```

```python
# Read csv file
Disney_file = "../Resources/Disney_Movies.csv"
Disney_df = pd.read_csv(Disney_file,index_col=0)
# making a copy with 'title'
movies_df= Disney_df[['title']].copy()
movies_df
```

```python
# create new columns
movies_df["budget"]=''
movies_df["revenue"]=''
movies_df["runtime"]=''
movies_df["movie_id"]=''
movies_df["genre"]=''
```

```python
# loop through the dataframe
for index,row in movies_df.iterrows():
# pass the movie title from the dataframe,fetch the movie id from the API
    movie_title=row['title']
    url = f"https://api.themoviedb.org/3/search/movie?api_key={api_key}&query={movie_title}"
    try:
        response = requests.get(url).json()
        movies_df.loc[index,'movie_id']=response['results'][0]['id']
    except (KeyError, IndexError, JSONDecodeError):
        print("Data not found... skipping.")
    except requests.ConnectionError:
        print("ConnectionError...")
```

```python
# loop through the sample dataframe
for index,row in movies_df.iterrows():
# pass the movie id and fetch the response from the API
        movie_id=row['movie_id']
        url = f"https://api.themoviedb.org/3/movie/{movie_id}?api_key={api_key}&language=en-US"
        try:
            response_id = requests.get(url).json()
            movies_df.loc[index,'budget']=response_id['budget']
            movies_df.loc[index,'revenue']=response_id['revenue']
            movies_df.loc[index,'runtime']=response_id['runtime']
            genre_type_list = response_id['genres'] 
            genre_type_str=""
            if genre_type_list != None :
                for genre_type in genre_type_list:
                    genre_type_str += genre_type['name']+","
            else:
                genre_type_str = "NA"
            movies_df.loc[index,"genre"]=genre_type_str 
    #         movie_id_list.append(movie_id)
        except (KeyError, IndexError, JSONDecodeError):
            print("Data not found... skipping.")
        except requests.ConnectionError:
            print("ConnectionError...")
```
