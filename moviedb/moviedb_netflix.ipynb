```python
#Dependencies
import pandas as pd
import requests
import json
from json.decoder import JSONDecodeError
from config import api_key
from pprint import pprint
```

```python
# Read csv file
netflix_file = "../Resources/Netflix_Movies.csv"
netflix_df = pd.read_csv(netflix_file,index_col=0)
# making a copy with 'title'
movies_df= netflix_df[['title']].copy()
movies_df.head()
```

```python
# create new columns
movies_df["budget"]=''
movies_df["revenue"]=''
movies_df["runtime"]=''
movies_df["movie_id"]=''
movies_df["genre"]=''


# take first 3131 records process and fetch API data,create a csv for first 3131 records
# movie_df=movies_df.head(3131)

#take next 3000 records process and fetcg API data,create a csv for the next 3000 records
movie_df=movies_df.tail(3000)
movie_df
```

```python
# loop through the sample dataframe
for index,row in movie_df.iterrows():
# pass the movie title from the dataframe,fetch the movie id from the API
    movie_title=row['title']
    url = f"https://api.themoviedb.org/3/search/movie?api_key={api_key}&query={movie_title}"
    try:
        print(index)
        response = requests.get(url).json()
        movie_df.loc[index,'movie_id']=response['results'][0]['id']
    except (KeyError, IndexError, JSONDecodeError):
        print("Data not found... skipping.")
    except requests.ConnectionError:
        print("ConnectionError...")
```

```python
# loop through the sample dataframe
for index,row in movie_df.iterrows():
# pass the movie id and fetch the response from the API
        movie_id=row['movie_id']
        url = f"https://api.themoviedb.org/3/movie/{movie_id}?api_key={api_key}&language=en-US"
        try:
            print(index)
            response_id = requests.get(url).json()
            movie_df.loc[index,'budget']=response_id['budget']
            movie_df.loc[index,'revenue']=response_id['revenue']
            movie_df.loc[index,'runtime']=response_id['runtime']
            genre_type_list = response_id['genres'] 
            genre_type_str=""
            if genre_type_list != None :
                for genre_type in genre_type_list:
                    genre_type_str += genre_type['name']+","
            else:
                genre_type_str = "NA"
            movie_df.loc[index,"genre"]=genre_type_str 
    #         movie_id_list.append(movie_id)
        except (KeyError, IndexError, JSONDecodeError):
            print("Data not found... skipping.")
        except requests.ConnectionError:
            print("ConnectionError...")
```

```python
# write the first 3131 and last 3000 in separate csv
# movie_df.to_csv('sample_head.csv')
movie_df.to_csv('sample_tail.csv')
```
