```python
import pandas as pd
```

```python
csv_file ='Resources/netflix_titles.csv'
data = pd.read_csv(csv_file,index_col=False)
data.head()
```

```python
filtered_data =data.loc[data['type']=='Movie']
filtered_data.reset_index(inplace=True)
```

```python
movies_data = filtered_data[['title','director','cast','country','release_year','rating']]
movies_data
```

```python
movies_data.to_csv('Resources/Netflix_Movies.csv')
```
