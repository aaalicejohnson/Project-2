```python
import pandas as pd
from sqlalchemy import create_engine
from sqlalchemy import inspect
import config

from config import user
from config import admin
```

```python
movies_file = 'moviedb/all_movies.csv'
all_movies_df = pd.read_csv(movies_file)
all_movies_df.head()
```

```python
all_movies_df = all_movies_df.drop(columns=["Unnamed: 0"])
all_movies_df.head()
```

```python
all_movies_df['id'] = all_movies_df.index
all_movies_df.head()
```

```python
all_movies_df.dtypes
```

```python
all_movies_df = all_movies_df.rename(columns={"cast": "movie_cast", "title": "title"})
all_movies_df.head()
```

```python
protocol = 'postgresql'
username = user
password = admin
host = 'localhost'
port = 5432
database_name = 'movies'

rds_connection_string = f'{protocol}://{username}:{password}@{host}:{port}/{database_name}'
        
engine = create_engine(rds_connection_string)
```

```python
inspector = inspect(engine)

inspector.get_table_names()
```

```python
all_movies_df.to_sql(name='all_movies', con=engine, if_exists='append', index=False)
```
